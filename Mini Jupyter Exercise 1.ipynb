### #1. Plot the GDP as a function of the year in Jupyter

```python
%matplotlib inline
import matplotlib.pyplot as plot
import numpy as numpy

Year = [1930 , 1940, 1950, 1960, 1970 , 1980 , 1990 , 2000  , 2010  ]
GDP  = [1.015, 1.33, 2.29, 3.26, 4.951, 6.759, 9.366, 13.131, 15.599] 

x = numpy.arange(9)
gdp = 1.823*x

plot.plot(Year, GDP, 'r--')
plot.scatter(Year, GDP, color = 'red')
plot.plot(Year, gdp, 'b', label = 'Predicted GDP')
plot.title('Real US GDP (in trillions)')
plot.xlabel('Year')
plot.ylabel('Trillions of $')
plot.legend(loc='upper left')
```

### #2: Based on the obtained plot, what is a possible mathematical relationship between GDP and Year? 

Below is a table illustrating the difference in GDP for a span of 10 years at a time. Based on the table, the average difference in GDP for a 10 year span is 1.823. 

| Difference in Years|'30 to '40|'40 to '50|'50 to '60|'60 to '70|'70 to '80|'80 to '90|'90 to '00|'00 to '10|
|--------------------|----------|----------|----------|----------|----------|----------|----------|----------|
| Difference in GDP  |   0.315  |   0.96   |   0.97   |   1.691  |   1.808  |   2.607  |   3.765  |   2.468  |

A possible mathematical relationship between GDP and year is that every year the GDP increases by 0.1823 trillions, or 182.3 billions, of dollars. I came to this conclusion by: $$Average \; Difference \; in \; GDP \; per \; Year = \frac{Average \; GDP \; Increase \; Every \; 10 \; Years \; (in \; trillions)}{10} = \frac{1.823 \; trillions}{10} = 0.1823 \; trillions $$

Although the actual GDP does not appear to increase linearly each year, I think my 0.1823 trillion increase per year is accurate because it averages out the larger growth during good economic times and the low growth during recessions and depressions.
